## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np

def simulacion_dados(n_lanzamientos=1000):
    conteo_exitos = 0
    
    for _ in range(n_lanzamientos):
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        
        total = dado1 + dado2
        
        if total > 7 or total % 2 == 0:
            conteo_exitos += 1
    
    probabilidad = conteo_exitos / n_lanzamientos
    return probabilidad
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
def extraer_bolas():
    bolas_extraidas = []
    for _ in range(5):
        bola_extraida = random.randint(0, 59)
        if bola_extraida < 10:
            bolas_extraidas.append("White")
        elif bola_extraida < 30:
            bolas_extraidas.append("Red")
        else:
            bolas_extraidas.append("Green")
    return bolas_extraidas

num_experimentos = 1000
conteo_3_blancas_2_rojas = 0
conteo_todas_mismo_color = 0

for _ in range(num_experimentos):
    bolas_extraidas = extraer_bolas()
    
    if bolas_extraidas.count("White") == 3 and bolas_extraidas.count("Red") == 2:
        conteo_3_blancas_2_rojas += 1
   
    if bolas_extraidas.count(bolas_extraidas[0]) == 5:
        conteo_todas_mismo_color += 1

probabilidad_3_blancas_2_rojas = conteo_3_blancas_2_rojas / num_experimentos
probabilidad_todas_mismo_color = conteo_todas_mismo_color / num_experimentos

print(f"Probabilidad de tomar 3 blancas y 2 rojas: {probabilidad_3_blancas_2_rojas:.4f}")
print(f"Probabilidad de tomar todas del mismo color: {probabilidad_todas_mismo_color:.4f}")
```
